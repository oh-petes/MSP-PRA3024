calculating kinetic energy for 1 million particles

```python
import numpy as np
```

```python
# define 1,000,000 particles
n_particles = 1000000

# Mass of each in particle in kg (will choose 0.5kg)
mass = 0.5

# generate 1,000,000 random velocities between 0 an 100 m/s
velocities = np.random.uniform(0, 100, n_particles)
```

```python
%%timeit
kinetic_energy = 0.5 * mass * velocities**2
```
